#### Housing Prices
Linear Regression builds a model of the relationship between features and targets. In this case, house size is the feature and house price is the target. For simple linear regression, the model has two parameters w and b, whose values are 'fit' using training data. Once we determine the model's parameters, we can use it to make predictions on new data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)

x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
# we can find this using the .shape property

print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is: {m}")
```

```python
# alternatively, we can use the len function to find the number of training examples

m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
# we can use (x^(i), y^(i)) to denote the ith example
# we start with i = 0 since Python is zero indexed

i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# now we can use matplotlib to plot the two points

plt.scatter(x_train, y_train, marker='x', c='r')
plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# lets calculate the predicted function output using a for loop

def compute_model_output(x, w, b):
        """
        Computes the prediction of a linear model
        Args:
            x (ndarray (m,)) - n dimentional array of shape m : Data, m examples
            w,b (scalar)                                      : model m parameters
        Returns
            y (ndarray (m,))                                  : target values
        """
        m = x.shape[0]
        f_wb = np.zeros(m)
        for i in range(m):
            f_wb[i] = w * x[i] + b
        
        return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b, )

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# clearly, w=100 and b=100 does not fit our data well
```

```python
# lets try again with new values for w and b
w = 200
b = 100

tmp_f_wb = compute_model_output(x_train, w, b, )

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title("Housing Prices")
plt.ylabel('Price (in 1000s of dollars)')
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```

```python
# Now, using our model we can predict the price of a house with 1200 sqft

w = 200
b = 100
x_i = 1.2
cost_1200sqft = w * x_i + b

print(f"${cost_1200sqft:.0f} thousand dollars")
```
